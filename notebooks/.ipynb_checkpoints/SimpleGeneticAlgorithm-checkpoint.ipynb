```julia
# custom result trace
function trace!(record::Dict{String,Any}, objfun, state, population, method, options)
    record["state"] = state
    record["objfun"] = objfun
    record["population"] = population
end
```

```julia
Evolutionary.initial_population(GA(populationSize = 12, selection = x->best(x,6),
            crossover = AX, mutation = PLM()),
            Evolutionary.Options(iterations=100,
                                 time_limit=8*60*60.0,
                                 store_trace = true,
                                 parallelization=:thread),BoxConstraints(lower, upper))
```

```julia
weights = [0.5, 0.5] # 50% on time, 50% on throttle
lower = [250.0e3, 0.0]
upper = [450.0e3, pi]
result = Evolutionary.optimize(
         x-> eval_fitness_val(x,weights),
         BoxConstraints(lower, upper),
         GA(populationSize = 12, selection = ranklinear(3),
            crossover = AX, mutation = PLM()),
            Evolutionary.Options(iterations=100,
                                 time_limit=8*60*60.0,
                                 store_trace = true,
                                 parallelization=:thread)) # :serial, :thread
```

```julia
opt_sc = spacecraft(result.minimizer[1], result.minimizer[2])
opt_traj = eval_fitness(opt_sc)
plotCOES(opt_traj)
```

```julia
using Plots
gene_hist1 = []
gene_hist2 = []
for iteration in saved.trace
    population = permutedims(hcat(iteration.metadata["population"]...))
    gene1 = population[:,1]
    gene2 = population[:,2]
    push!(gene_hist1, gene1)
    push!(gene_hist2, gene2)
end
gene_hist1

splt = scatter(ones(size(gene_hist1)), gene_hist1[1])
for i = 2:length(gene_hist1)
    scatter!(splt, i.*ones(size(gene_hist1)), gene_hist1[i])
end
@show splt
```
